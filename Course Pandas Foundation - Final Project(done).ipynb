```python
# TODO 05 - how much total sales, average sales, and standard deviation of sales your company make in 2017
import pandas as pd

df = pd.read_csv('sample-store.csv')

df['Order Date'] = pd.to_datetime(df['Order Date'], format = '%m/%d/%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format = '%m/%d/%Y')

# filer year = 2017
df_2017 = df.loc[df['Order Date'].dt.year == 2017]

# Aggregrate function
totol_sales = df_2017['Sales'].sum()
avg_sales = df_2017['Sales'].mean()
std_sales = df_2017['Sales'].std()

# Print result
print(totol_sales, avg_sales, std_sales)
```

```python
# TODO 06 - which Segment has the highest profit in 2018
import pandas as pd

df = pd.read_csv('sample-store.csv')

df['Order Date'] = pd.to_datetime(df['Order Date'], format = '%m/%d/%Y')

df_2018 = df.loc[df['Order Date'].dt.year == 2018]

# find Highest profit groupby Segment
Segment_Profit = df_2018.groupby('Segment')['Profit'].sum()

Segment_highest = Segment_Profit.idxmax()
highest_profit = Segment_Profit.max()
print(Segment_Profit)

# แสดงออกมาได้หลายแบบ
print(df_2018.groupby('Segment')['Profit'].sum().idxmax())
print(df_2018.groupby('Segment')['Profit'].sum().max())

print(f"{Segment_highest} : {highest_profit :,.2f}") # :,.2f คือการแสดง float 2 ตำแหน่ง
```

```python
# TODO 07 - which top 5 States have the least total sales between 15 April 2019 - 31 December 2019
import pandas as pd

df = pd.read_csv('sample-store.csv')

df['Order Date'] = pd.to_datetime(df['Order Date'], format = '%m/%d/%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format = '%m/%d/%Y')

# filter datetime between 15 Apr - 31 Dec 2019
date = (df['Order Date'] >= '2019-04-15') & (df['Order Date'] <= '2019-12-31')
df_prep_date = df[date]

# find the solution
sales_by_state = (
    df_prep_date.groupby('State')['Sales'].sum()
    .sort_values( ascending = True)
    .head(5)
)

print(sales_by_state)
```

```python
# TODO 08 - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25% 
import pandas as pd

df = pd.read_csv('sample-store.csv')

df['Order Date'] = pd.to_datetime(df['Order Date'], format = '%m/%d/%Y')

# filter date
date = df['Order Date'].dt.year == 2019

total_sales_2019 = df.loc[date, 'Sales'].sum()
print(f"total sales of all Region is : {total_sales_2019:,.2f}")



# find sales
sum_west_cen = df[date].query( 'Region in ["West", "Central"]' )['Sales'].sum()
print(f"total sales of West + Central is : {sum_west_cen:,.2f}")

print(f"Proportion is {(sum_west_cen / total_sales_2019) * 100:,.2f}%")
```

```python
# TODO 09 - find top 10 popular products in terms of number of orders vs. total sales during 2019-2020
import pandas as pd

df = pd.read_csv('sample-store.csv')
df['Order Date'] = pd.to_datetime(df['Order Date'], format = '%m/%d/%Y')

# filter year
df_19_20 = df[ ( df['Order Date'].dt.year.isin([2019, 2020]) ) ]

# check result
print(df_19_20['Order Date'].dt.year.unique())

# df_19_20.groupby('Sub-Category')['Sales'].sum().sort_values(ascending = False).head(10)
product_summary = df_19_20.groupby('Product Name').agg(
    Order_Count = ('Order ID', 'count'),
    total_sales = ('Sales', 'sum'),
)
top10_orders = product_summary.sort_values(by = "Order_Count", ascending = False).head(10)
print(top10_orders)

top10_total_sales = product_summary.sort_values(by = "total_sales", ascending = False).head(10)
print(top10_total_sales)
```

```python
# TODO 10 - plot at least 2 plots, any plot you think interesting :)

import pandas as pd

df = pd.read_csv('sample-store.csv')

df['Order Date'] = pd.to_datetime(df['Order Date'], format = '%m/%d/%Y')

df_2019 = df[df['Order Date'].dt.year == 2019]
sales_2019_region = df_2019.groupby('Region')['Sales'].sum()
print(sales_2019_region.round(2))

import matplotlib.pyplot as plt

plt.figure(figsize = (7, 7))
plt.pie(
    sales_2019_region,
    labels = sales_2019_region.index,
    autopct = '%0.2f%%',
    pctdistance = 0.75, # ค่าน้อย = ใกล้ศูนย์กลาง, ค่ามาก = ขยับออก
    startangle = 140,
    wedgeprops = {'width': 0.5} # ตรงนี้ ทำให้เป็น donut chart
)
plt.title('Sales Proportion by Region (Donut Chart)')
plt.axis('equal') # ทำให้เป็นวงกลมๆ
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

df = pd.read_csv('sample-store.csv')

df['Order Date'] = pd.to_datetime(df['Order Date'], format = '%m/%d/%Y')

df_2019 = df[df['Order Date'].dt.year == 2019].copy()
df_2019['Month'] = df_2019['Order Date'].dt.to_period('M')
Monthly_Sales = df_2019.groupby('Month')['Sales'].sum()

import matplotlib.pyplot as plt

Monthly_Sales.plot(kind = 'line', marker = 'o')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Monthly Sales Trend (2019)')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```
